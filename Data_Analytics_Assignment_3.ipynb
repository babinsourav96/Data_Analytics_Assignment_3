1. What is the difference between a function and a method in Python?

  - Association: Functions are independent; methods are associated with classes/objects.
  - Calling: Functions are called by name; methods are called on an object using dot notation (object.method()).
  - Implicit Argument: Methods implicitly receive the object instance as their first argument (self), while functions do not have this implicit argument.
  - Purpose: Functions are for general tasks; methods define behavior specific to an object's state and actions.
  

```python
#Function example:-
def greet(name):
    """This is a simple function that greets a person."""
    print(f"Hello, {name}!")

greet("Alice") # Calling the function
```

```python
#Method Example:-
class Dog:
    """This is a simple class representing a Dog."""
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        """This is a method that makes the dog bark."""
        print(f"{self.name} says Woof!")

my_dog = Dog("Buddy", "Golden Retriever") # Creating an object (instance of the class)
my_dog.bark() # Calling the method on the object
```

2. Explain the concept of function arguments and parameters in Python.

  - Arguments are the actual values or data passed to a function when it is called or invoked. These values are then assigned to the corresponding parameters within the function's scope.
  - Parameters are placeholders for values defined in the function's signature or definition. They are the variables listed inside the parentheses in the def statement when you create a function.

```python
#Arguments example:-
greet("Alice")  # "Alice" are arguments
```

```python
#Parameters example:-
def greet(name, age):  # 'name' and 'age' are parameters
    """
    This function takes a name and age and prints a greeting.
    """
    print(f"Hello, {name}! You are {age} years old.")

# Calling the function with arguments
greet("Alice", 30)  # "Alice" and 30 are arguments
greet("Bob", 25)    # "Bob" and 25 are arguments
```

4. What is the purpose of the `return` statement in a Python function?

  - The return statement in a Python function serves two primary purposes:
    - Exiting the Function: When a return statement is encountered, the function's execution immediately terminates, and control is transferred back to the point in the code where the function was called. Any code within the function after the return statement will not be executed.
    - Returning a Value: The return statement can optionally be followed by an expression or a value. This value is then sent back to the caller of the function and can be stored in a variable, used in further calculations, or passed to other functions. If no value or expression is specified after return, or if there is no return statement in the function, the function implicitly returns None.

```python
#Example:-
def calculate_area(length, width):
  """
  Calculates the area of a rectangle.
  """
  area = length * width
  return area  # Returns the calculated area

def greet(name):
  """
  Prints a greeting message.
  """
  print(f"Hello, {name}!")
  # No return statement, so it implicitly returns None

# Calling calculate_area and storing the returned value
rectangle_area = calculate_area(5, 10)
print(f"The area of the rectangle is: {rectangle_area}")

# Calling greet (it doesn't return a value to be stored)
greet("Alice")

# Demonstrating implicit None return
result_of_greet = greet("Bob")
print(f"The return value of greet is: {result_of_greet}")
```

5. What are iterators in Python and how do they differ from iterables?

  - An iterator is an object that represents a stream of data and provides a way to access elements one at a time.

  - Difference between Iterators and Iterables:-
    - Ability to be iterated over: Iterables can be iterated over directly (e.g., in a for loop), while iterators are the objects that perform the actual iteration.
    - Methods implemented: Iterables implement __iter__(), while iterators implement both __iter__() and __next__().
    - Statefulness:Iterators maintain an internal state to keep track of the current position during iteration, allowing them to return the "next" item. Iterables typically do not maintain this state directly.
    - Relationship: Every iterator is also an iterable, as it implements __iter__(). However, not every iterable is an iterator. For example, a list is an iterable, but it's not an iterator itself; you get an iterator from the list using iter().

6. Explain the concept of generators in Python and how they are defined.

  - A generator function is defined like a regular function, but instead of using return, it uses the yield keyword to produce values. When a generator function is called, it doesn't execute its code immediately.
  

```python
#Example:-

def even_numbers(max_val):
    n = 0
    while n <= max_val:
        if n % 2 == 0:
            yield n
        n += 1

# Using the generator
for num in even_numbers(10):
    print(num)

# Or, to get a list
even_list = list(even_numbers(10))
print(even_list)
```

7. What are the advantages of using generators over regular functions?

  - Advantages are as follows:-
    - Memory Efficiency (Lazy Evaluation): Generators compute and yield values one at a time, on demand, rather than generating and storing an entire sequence in memory at once.
    - Handling Infinite Sequences: Because generators produce values on demand, they can easily represent and work with infinite sequences, which is impossible with regular functions that must return a complete collection.
    - Simplified Iterative Logic: Generators simplify the creation of iterators, as they automatically handle the state management involved in pausing and resuming execution, eliminating the need to manually implement __iter__ and __next__ methods.
    - Improved Readability and Conciseness: The yield keyword makes the code for generating sequences more intuitive and expressive, often leading to cleaner and more readable code compared to managing explicit state variables in a regular function.

8.  What is a lambda function in Python and when is it typically used?

  - In Python, a lambda function is a small, anonymous function defined using the lambda keyword. It can take any number of arguments but can only have one expression.Lambda functions are typically used for concise, single-line operations and are often passed as arguments to higher-order functions like map(), filter(), and sorted().


```python
#Example:-
# A lambda function that adds 10 to its argument
add_10 = lambda x: x + 10

# Calling the lambda function
result = add_10(5)
print(result)  # Output: 15

# Using a lambda function with map() to square each element in a list
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
```

9. Explain the purpose and usage of the `map()` function in Python.

  - The primary purpose of map() is to:
        - Transform Data: Apply a specific operation or transformation to every element within an iterable.
        - Avoid Explicit Loops: Offer a more functional programming approach by eliminating the need for explicit for loops, leading to cleaner and potentially more performant code, especially for large datasets.
        - Process Iterables Efficiently: Return an iterator (a map object in Python 3.x), which generates results on demand rather than creating a new list in memory immediately, thus optimizing memory usage.
  - Usage:-
        - function: The function to be applied to each item of the iterable. This function can be a built-in function, a user-defined function, or a lambda function.
        - iterable: One or more iterable objects whose elements will be passed as arguments to the function. If multiple iterables are provided, the function must accept a corresponding number of arguments.

```python
#Example:-

# Define a function to square a number
def square(n):
    return n * n

numbers = [1, 2, 3, 4, 5]

# Use map() to apply the square function to each number
squared_numbers_map = map(square, numbers)

# Convert the map object to a list to view the results
squared_numbers_list = list(squared_numbers_map)
print(squared_numbers_list)

# Using a lambda function with map()
names = ["alice", "bob", "charlie"]
capitalized_names = list(map(lambda name: name.upper(), names))
print(capitalized_names)
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

  - map(function, iterable): Applies a given function to each item in an iterable and returns an iterator that yields the results.
  Example: Doubling each number in a list.
  - filter(function, iterable): Constructs an iterator from elements of an iterable for which a given function returns True.
  Example: Extracting only even numbers from a list.
  - reduce(function, iterable, initializer): Applies a given function cumulatively to the items of an iterable, reducing the iterable to a single cumulative value. It requires importing from the functools module (from functools import reduce).
  Example: Calculating the sum of all numbers in a list.

```python
# Write a Python function that checks if a given number is prime or not from 1 to 200.

import math

def is_prime(num):
    if num <= 1:
        return False  # Numbers less than or equal to 1 are not prime
    if num == 2:
        return True   # 2 is the only even prime number
    if num % 2 == 0:
        return False  # Other even numbers are not prime

    # Check for divisibility by odd numbers up to the square root of num
    for i in range(3, int(math.sqrt(num)) + 1, 2):
        if num % i == 0:
            return False
    return True

def find_primes_in_range(start, end):
    print(f"Prime numbers between {start} and {end}:")
    for number in range(start, end + 1):
        if is_prime(number):
            print(number)

# Example usage for numbers from 1 to 200
find_primes_in_range(1, 200)
```

```python
#  Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
with open("sample.txt", "w") as f:
    f.write("First line of text.\n")
    f.write("Second line here.\n")
    f.write("Third and final line.")
for line in read_file_line_by_line("sample.txt"):
    print(line)
```

```python
# Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
  return (celsius * 9/5) + 32
def convert_temperatures(celsius_list):
    fahrenheit_list = list(map(celsius_to_fahrenheit, celsius_list))
    return fahrenheit_list
celsius_temperatures = [0, 10, 20, 30, 40]
fahrenheit_temperatures = convert_temperatures(celsius_temperatures)
print(f"Celsius: {celsius_temperatures}")
print(f"Fahrenheit: {fahrenheit_temperatures}")
```
